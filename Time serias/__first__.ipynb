```python
from sklearn import linear_model
import datetime
import time
```

```python
def county_predict(number_of_country_in_dataset, date_to_predict):
    # date_to_predict convert to timestamp
    date_to_predict = np.array(time.mktime(datetime.datetime.strptime(date_to_predict + '20', "%m/%d/%Y").timetuple()))
    date_to_predict = date_to_predict.reshape(-1, 1)
    # date 
    date_time = []
    date_str = np.array(df.columns)[11:]
    for i in date_str:
        d = time.mktime(datetime.datetime.strptime(i + '20', "%m/%d/%Y").timetuple())
        date_time.append(d)

    # y = the number of death people in day
    number_of_death = []
    data = df.iloc[number_of_country_in_dataset, 11:]
    for i in data:
        number_of_death.append(i)
        
    # reshape
    date_time = np.array(date_time).reshape(-1, 1)
    # predict to other day using linear model
    regr = linear_model.LinearRegression()
    regr.fit(date_time, number_of_death)
    # predict
    y_pred = regr.predict(date_to_predict)
    #plot
    plt.figure(figsize=(20, 7))
    plt.title('The number of death in ' + str(df['Country'][number_of_country_in_dataset]))
    plt.xlabel('Date')
    plt.ylabel('Metric')
    
    plt.scatter(date_time, number_of_death,  color='blue')
    plt.scatter(date_to_predict, y_pred, color='red', linewidth=3)
    print(date_to_predict, y_pred)
    plt.show()
```

```python
county_predict(132, '08/11/20')
```

#### Коинтеграция

```python
import statsmodels.tsa.stattools as ts
```

```python
def coint(country_1, country_2):
    # date_index
    date_index = []
    
    date_str = df.iloc[country_1, 11:].index
    for i in date_str:
        d = time.mktime(datetime.datetime.strptime(i + '20', "%m/%d/%Y").timetuple())
        date_index.append(d)
    
    data1 = df.iloc[country_1, 11:]
    data2 = df.iloc[country_2, 11:]
    
    coint_result = ts.coint(data1, data2)
    
    return coint_result
```

Первым, кто ввел и проанализировал концепцию ложной или бессмысленной регрессии, был Удни Йоль в 1926 году. До 1980-х годов многие экономисты использовали линейные регрессии на данных нестационарных временных рядов, которые показали нобелевские лауреаты Клайв Грейнджер и Пол Ньюболд. быть опасным подходом, который может привести к ложной корреляции , поскольку стандартные методы убывания могут привести к тому, что данные все еще будут нестационарными.

```python
for i in range(len(df)):
    print("==============================================================================")
    for j in range(len(df)):
        if i != j:
            print(f"{df['Country'][i]} coint {df['Country'][j]} result = {coint(i, j)[1]}")
```
