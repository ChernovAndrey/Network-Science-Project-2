```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
deaths = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
population = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv'
country_population='https://raw.githubusercontent.com/DmitrySerg/COVID-19/master/data/country_population.csv'
airports = 'https://raw.githubusercontent.com/DmitrySerg/COVID-19/master/data/airport_df.csv'
```

```python
e1_incomesupport = 'https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/timeseries/e1_incomesupport.csv'
e2_debtcontractrelief = 'https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/timeseries/e2_debtcontractrelief.csv'
h1_publicinfocampaign = 'https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/timeseries/h1_publicinfocampaign.csv'
h2_testingpolicy = 'https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/timeseries/h2_testingpolicy.csv'
h3_contacttracing = 'https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/timeseries/h3_contacttracing.csv'
```

```python
df = pd.read_csv(confirmed)
for country in ['Australia', 'Canada', 'China', 'Denmark', 'France', 'Netherlands', 'United Kingdom']:
    df_country = df.loc[df['Country/Region'] == country]
    df_country = df_country.drop(['Province/State', 'Country/Region'], axis=1)
    ser_country = df_country.sum()
    ser_country[['Lat', 'Long']] /= len(df_country)
    ser_country['Country/Region'] = country
    df = df.drop(df_country.index).append(ser_country, ignore_index=True)
df_confirmed = df.drop('Province/State', axis=1).rename(columns={'Country/Region': 'Country'})
df_confirmed.loc[df_confirmed['Country'] == 'US', 'Country'] = 'United States'
df_confirmed.loc[df_confirmed['Country'] == 'Korea, South', 'Country'] = 'South Korea'
df_confirmed.loc[df_confirmed['Country'] == 'Taiwan*', 'Country'] = 'Taiwan'
df_confirmed.loc[df_confirmed['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
df_confirmed.loc[df_confirmed['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'
```

```python
df = pd.read_csv(deaths)
for country in ['Australia', 'Canada', 'China', 'Denmark', 'France', 'Netherlands', 'United Kingdom']:
    df_country = df.loc[df['Country/Region'] == country]
    df_country = df_country.drop(['Province/State', 'Country/Region'], axis=1)
    ser_country = df_country.sum()
    ser_country[['Lat', 'Long']] /= len(df_country)
    ser_country['Country/Region'] = country
    df = df.drop(df_country.index).append(ser_country, ignore_index=True)
df_deaths = df.drop('Province/State', axis=1).rename(columns={'Country/Region': 'Country'})
df_deaths.loc[df_deaths['Country'] == 'US', 'Country'] = 'United States'
df_deaths.loc[df_deaths['Country'] == 'Korea, South', 'Country'] = 'South Korea'
df_deaths.loc[df_deaths['Country'] == 'Taiwan*', 'Country'] = 'Taiwan'
df_deaths.loc[df_deaths['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
df_deaths.loc[df_deaths['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'
```

```python
df = pd.read_csv(recovered)
for country in ['Australia', 'Canada', 'China', 'Denmark', 'France', 'Netherlands', 'United Kingdom']:
    df_country = df.loc[df['Country/Region'] == country]
    df_country = df_country.drop(['Province/State', 'Country/Region'], axis=1)
    ser_country = df_country.sum()
    ser_country[['Lat', 'Long']] /= len(df_country)
    ser_country['Country/Region'] = country
    df = df.drop(df_country.index).append(ser_country, ignore_index=True)
df_recovered = df.drop('Province/State', axis=1).rename(columns={'Country/Region': 'Country'})
df_recovered.loc[df_recovered['Country'] == 'US', 'Country'] = 'United States'
df_recovered.loc[df_recovered['Country'] == 'Korea, South', 'Country'] = 'South Korea'
df_recovered.loc[df_recovered['Country'] == 'Taiwan*', 'Country'] = 'Taiwan'
df_recovered.loc[df_recovered['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
df_recovered.loc[df_recovered['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'
```

```python
df = pd.read_csv(population)
df = df.loc[df['Province_State'].isna()]
df_population = df[['Country_Region', 'Population']]
df_population = df_population.rename(columns={'Country_Region': 'Country'})
df_population.loc[df_population['Country'] == 'US', 'Country'] = 'United States'
df_population.loc[df_population['Country'] == 'Korea, South', 'Country'] = 'South Korea'
df_population.loc[df_population['Country'] == 'Taiwan*', 'Country'] = 'Taiwan'
df_population.loc[df_population['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
df_population.loc[df_population['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'
```

```python
df = pd.read_csv(country_population)
df_area = df[['name', 'area', 'Density']]
df_area = df_area.rename(columns={'name': 'Country', 'area': 'Area'})
df_area.loc[df_area['Country'] == 'DR Congo', 'Country'] = 'Democratic Republic of the Congo'
```

```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
results=soup('tr', valign='top')
results=results[0].find('td')
results=results.find_all('tr')
data=[]
for result in results[2:]:
    res=result.find_all('td')
    country = res[1].a['title'].strip()
    gdp = int(res[2].text.strip().replace(',','').replace('/Na',''))
    data.append((country, gdp))
df_gdp = pd.DataFrame(data, columns=['Country', 'GDP'])
df_gdp.loc[df_gdp['Country'] == 'The Bahamas', 'Country'] = 'Bahamas'
df_gdp.loc[df_gdp['Country'] == 'The Gambia', 'Country'] = 'Gambia'
```

```python
df = pd.read_csv(airports)
df_airports = df.groupby('Country').count()['ID'].reset_index()
df_airports = df_airports.rename(columns={'ID': 'Airports'})
df_airports.loc[df_airports['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
df_airports.loc[df_airports['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'
```

```python
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
df_owid = df.groupby('location')[['handwashing_facilities', 'hospital_beds_per_100k']].max().reset_index()
df_owid = df_owid.rename(columns={'location': 'Country'})
df_owid.loc[df_owid['Country'] == 'Democratic Republic of Congo', 'Country'] = 'Democratic Republic of the Congo'
df_owid.loc[df_owid['Country'] == 'Congo', 'Country'] = 'Republic of the Congo'
```

```python
df1 = pd.read_csv(c1_schoolclosing)
df2 = pd.read_csv(c2_workplaceclosing)
df3 = pd.read_csv(c3_cancelpublicevents)
df4 = pd.read_csv(c4_restrictionsongatherings)
df5 = pd.read_csv(c5_closepublictransport)
df6 = pd.read_csv(c6_stayathomerequirements)
df7 = pd.read_csv(c7_domestictravel)
df8 = pd.read_csv(c8_internationaltravel)
df9 = pd.read_csv(e1_incomesupport)
df10 = pd.read_csv(e2_debtcontractrelief)
df11 = pd.read_csv(h1_publicinfocampaign)
df12 = pd.read_csv(h2_testingpolicy)
df13 = pd.read_csv(h3_contacttracing)
```

```python
def change_data(data):
    day = data[:2]
    month = data[2:-4]
    year = data[-2:]
    if month == 'jan':
        month = '1'
    elif month == 'feb':
        month = '2'
    elif month == 'mar':
        month = '2'
    elif month == 'apr':
        month = '3'
    elif month == 'may':
        month = '4'
    elif month == 'jun':
        month = '5'
    return '{}/{}/{}'.format(month, day, year)

def change_data_1(data):
    year, month, day = data.split('-')
    if month.startswith('0'):
        month = month[1:]
    year = year[-2:]
    return '{}/{}/{}'.format(month, day, year)
```

```python
dfs = [df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13]
countries = df1['Unnamed: 0']
data = df1.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
columns = data.columns
new_columns = [change_data(i) for i in columns]
dic = dict()
for i in range(len(countries)):
    dic[i] = np.array(data.iloc[i].apply(list).values)
for df in dfs:
    data = df.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
    for i in range(len(countries)):
        dic[i] += np.array(data.iloc[i].apply(list).values)
df_policy = pd.DataFrame(dic, index=new_columns).T.iloc[:, 21:]
df_policy.insert(0, 'Country', countries)
df_policy.loc[df_airports['Country'] == 'Congo', 'Country'] = 'Republic of the Congo'
df_policy = df_policy.drop([170, 171, 172])
```

```python
df = pd.read_csv('https://www.gstatic.com/covid19/mobility/Global_Mobility_Report.csv?cachebust=b3ad67084527db32')
df = df.loc[df['sub_region_1'].isna()]
df = df.drop(['country_region_code', 'sub_region_1', 'sub_region_2'], axis=1)
countries = pd.value_counts(df['country_region']).index
dic = dict()
for num, country in enumerate(countries):
    df_mini = df.loc[df['country_region'] == country]
    ind = []
    for i in range(len(df_mini)):
        ind.append([df_mini.iloc[i, j + 2] for j in range(6)])
    dic[num] = ind
columns = df_mini['date']
new_columns = [change_data_1(i) for i in columns]
df_mobility = pd.DataFrame(dic, index=new_columns).T
df_mobility.insert(0, 'Country', countries)
df_mobility.loc[df_mobility['Country'] == 'The Bahamas', 'Country'] = 'Bahamas'
df_mobility = df_mobility.sort_values(['Country'])
```

```python
df_confirmed_0 = pd.merge(df_airports, df_confirmed, how='right', on='Country')
df_confirmed_0 = pd.merge(df_owid, df_confirmed_0,  how='right', on='Country')
df_confirmed_0 = pd.merge(df_gdp, df_confirmed_0,  how='right', on='Country')
df_confirmed_0 = pd.merge(df_area, df_confirmed_0,  how='right', on='Country')
df_confirmed_0 = pd.merge(df_population, df_confirmed_0,  how='right', on='Country')
df_confirmed_0 = df_confirmed_0.sort_values(['Country'])
df_confirmed_0
```

```python
df_deaths_0 = pd.merge(df_airports, df_deaths, how='right', on='Country')
df_deaths_0 = pd.merge(df_owid, df_deaths_0,  how='right', on='Country')
df_deaths_0 = pd.merge(df_gdp, df_deaths_0,  how='right', on='Country')
df_deaths_0 = pd.merge(df_area, df_deaths_0,  how='right', on='Country')
df_deaths_0 = pd.merge(df_population, df_deaths_0,  how='right', on='Country')
df_deaths_0 = df_deaths_0.sort_values(['Country'])
df_deaths_0
```

```python
df_recovered_0 = pd.merge(df_airports, df_recovered, how='right', on='Country')
df_recovered_0 = pd.merge(df_owid, df_recovered_0,  how='right', on='Country')
df_recovered_0 = pd.merge(df_gdp, df_recovered_0,  how='right', on='Country')
df_recovered_0 = pd.merge(df_area, df_recovered_0,  how='right', on='Country')
df_recovered_0 = pd.merge(df_population, df_recovered_0,  how='right', on='Country')
df_recovered_0 = df_recovered_0.sort_values(['Country'])
df_recovered_0
```

```python
df_confirmed_0.to_csv('confirmed.csv')
df_deaths_0.to_csv('deaths.csv')
df_recovered_0.to_csv('recovered.csv')
df_policy.to_csv('policy.csv')
df_mobility.to_csv('mobility.csv')
```
